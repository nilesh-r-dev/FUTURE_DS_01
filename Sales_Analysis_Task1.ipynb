```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    'Order Date': ['2024-01-05','2024-01-15','2024-02-10','2024-02-18',
                   '2024-03-05','2024-03-12','2024-04-01','2024-04-15'],
    'Product': ['Laptop','Phone','Laptop','Tablet',
                'Phone','Tablet','Laptop','Phone'],
    'Category': ['Electronics','Electronics','Electronics','Electronics',
                 'Electronics','Electronics','Electronics','Electronics'],
    'Region': ['North','South','West','East',
               'North','West','South','East'],
    'Quantity': [2,3,1,4,2,1,3,2],
    'Sales': [120000,60000,60000,80000,40000,20000,180000,50000]
}

df = pd.DataFrame(data)
df
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
df.info()
```

```python
total_sales = df['Sales'].sum()
print("Total Sales:", total_sales)
```

```python
sales_by_product = df.groupby('Product')['Sales'].sum()
sales_by_product
```

```python
sales_by_region = df.groupby('Region')['Sales'].sum()
sales_by_region
```

```python
sales_by_product.plot(kind='bar')
plt.title("Sales by Product")
plt.ylabel("Sales Amount")
plt.show()
```

```python
sales_by_product.plot(kind='pie')
plt.title("Sales by Product")
plt.ylabel("Sales Amount")
plt.show()
```

```python
monthly_sales = df.resample('M', on='Order Date')['Sales'].sum()

monthly_sales.plot()
plt.title("Monthly Sales Trend")
plt.ylabel("Sales")
plt.show()
```
